```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("anime.csv")
```

```python
df1=pd.read_csv("anime_with_synopsis.csv")
```

```python
df1.drop("Name",axis=1,inplace=True)
df1.drop("Score",axis=1,inplace=True)
df1.drop("Genres",axis=1,inplace=True)
```

```python
df.columns
```

```python
df.drop("English name",axis=1,inplace=True)
df.drop("Producers",axis=1,inplace=True)
df.drop("Licensors",axis=1,inplace=True)
df.drop("Source",axis=1,inplace=True)
df.drop("Members",axis=1,inplace=True)
df.drop("Favorites",axis=1,inplace=True)
df.drop("Watching",axis=1,inplace=True)
df.drop("Completed",axis=1,inplace=True)
df.drop(["On-Hold", 'Dropped','Plan to Watch', 'Score-10', 'Score-9', 'Score-8', 'Score-7', 'Score-6','Score-5', 'Score-4', 'Score-3', 'Score-2', 'Score-1'],axis=1,inplace=True)
```

```python
df.head(3)
```

```python
df.columns
```

```python
df2=df1.merge(df,on='MAL_ID')
```

```python
df2.head(4)
```

```python
df2.columns
```

```python
df2.info()
```

```python
df2['Popularity']=df2['Popularity'].astype(str)
```

```python
df2.drop_duplicates(subset ="Name", keep = 'last', inplace = True)
```

```python
df2.iloc[0].Genres
```

```python
df2.dropna(inplace=True)
```

```python
df2.head()
```

```python
df2.duplicated().sum()
```

```python
df2['sypnopsis'] = df2['sypnopsis'].apply(lambda x:x.split())
```

```python
df2.head()
```

```python
df2['Genres']=df2['Genres'].apply(lambda x:x.split())
```

```python
df2['Studios']=df2['Studios'].apply(lambda x:x.split())
```

```python
df2.head()
```

```python
df2['Type']=df2['Type'].apply(lambda x:x.split())
```

```python
df2['tags']=df2['sypnopsis']+df2['Genres']+df2['Studios']+df2['Type']
```

```python
new = df2.drop(columns=['sypnopsis','Genres','Studios','Type'])
```

```python
new['tags'] = new['tags'].apply(lambda x: " ".join(x))
new.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
    
```

```python

vector = cv.fit_transform(new['tags']).toarray()
 
```

```python

vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity
```

```python
new[new['Name'] == 'Cowboy Bebop: Tengoku no Tobira'].index[0]
```

```python

def recommend(movie):
    index = new[new['Name'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(new.iloc[i[0]].Name)
        
```

```python
recommend('One Piece')
```
